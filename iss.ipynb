```python
#Collecting Data Points

import requests
import time

def get_iss_location():
    response = requests.get('http://api.open-notify.org/iss-now.json')
    data = response.json()
    return data['iss_position']['latitude'], data['iss_position']['longitude'], data['timestamp']
```

```python
data_points = []
for i in range(100):
    latitude, longitude, timestamp = get_iss_location()
    data_points.append((latitude, longitude, timestamp))
    time.sleep(10)
```

```python
#Visualization of Collected Points

import matplotlib.pyplot as plt

latitudes = [data[0] for data in data_points]
longitudes = [data[1] for data in data_points]

plt.scatter(longitudes, latitudes)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Path of the International Space Station')
plt.show()
```
